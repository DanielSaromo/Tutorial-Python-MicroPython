### M√©todo 1: Monte Carlo

```python
import random
def generarNvalores(n, a, b):
    """Genera una lista de `n` valores, que van desde `a` hasta `b` (intervalo cerrado)."""
    x = []

    for _ in range(n):

        #generate the n-th sample and append it to x
        x.append(random.uniform(a, b)) #https://docs.python.org/3/library/random.html#random.uniform

    return x

cant_puntos = 50000
x = generarNvalores(cant_puntos, 0, 4)
y = generarNvalores(cant_puntos, 0, 4)
```

```python
import numpy as np
def efe(x):
    r = 4
    #return np.power(x,2)#
    return np.sqrt(np.power(r,2) - np.power(x,2))

#y = efe(x)
efe(x).shape
```

```python
def cuentaPuntosDebajoDeEfe(x, y, f):
    return np.where(y<=f(x), 1, 0)
```

```python
vector_isBelow = cuentaPuntosDebajoDeEfe(x, y, f=efe)
A1 = vector_isBelow.sum()
```

```python
def integral_monteCarlo(x_a, x_b, y_a, y_b, f, n=5000):
  
    x = generarNvalores(n, 0, 4)
    y = generarNvalores(n, 0, 4)

    vector_isBelow = cuentaPuntosDebajoDeEfe(x, y, f=efe)
    A1 = vector_isBelow.sum()
    A1masA2 = len(x)

    areaRectangulo = (y_b - y_a) * (x_b - x_a)

    return A1/A1masA2 * areaRectangulo
```

### M√©todo 2: 1MAT07 / MAT129

```python
def integral_sumRiemann(f, a, b, n=5000):
    '''Calculates the integral based on the right-sided the Riemann Sums.
    
    :param function f: the integrand function
    :param int a: lower bound of the integral
    :param int b: upper bound of theintergal
    :param int n: number of rectangles of equal width
    :return float: the integral of the function f between a and b
    '''
    w = (b - a)/n
    result = sum([f(a + i*w) for i in range(1, n)])
    result *= w
    return result
```

### M√©todo 3: Riemann trapezoidal

```python
def integral_trapezoide(f, a, b, n=5000): #https://en.wikipedia.org/wiki/Trapezoidal_rule
    '''Calculates the integral based on the composite trapezoidal rule
    relying on the Riemann Sums.
    
    :param function f: the integrand function
    :param int a: lower bound of the integral
    :param int b: upper bound of theintergal
    :param int n: number of trapezoids of equal width
    :return float: the integral of the function f between a and b
    '''
    w = (b - a)/n
    result = 0.5*f(a) + sum([f(a + i*w) for i in range(1, n)]) + 0.5*f(b)
    result *= w
    return result
```

```python
class Rectangulo(object):
    def __init__(self, alto, ancho):
        self.alto = alto
        self.ancho = ancho
        self.area = 0
    
    def __repr__(self):
        return "Rectangulito(alto=%s, ancho=%s)"%(self.alto, self.ancho)

    def calcula_area(self):
        return self.alto * self.ancho

class Cuadrado(Rectangulo):
    def __init__(self, lado):
        self.alto = lado
        self.ancho = lado
```
